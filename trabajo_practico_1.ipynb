```python
#ejercicio 8
import numpy as np
import matplotlib.pyplot as plt

# Configuración general de subplots
fig, axs = plt.subplots(3, 3, figsize=(15, 10))
fig.suptitle("Verificación gráfica de límites notables", fontsize=16)

x = np.linspace(-0.1, 0.1, 1000)
x_pos = np.linspace(0.01, 10, 1000)
x_inf = np.linspace(1, 1000, 1000)

# Lim 1: lim x→0 (sen(x)/x) = 1
axs[0, 0].plot(x, np.sin(x)/x)
axs[0, 0].axhline(1, color='red', linestyle='--')
axs[0, 0].set_title(r'$\lim_{x \to 0} \frac{\sin x}{x} = 1$')

# Lim 2: lim x→0 (1-cos(x))/x = 0
axs[0, 1].plot(x, (1 - np.cos(x))/x)
axs[0, 1].axhline(0, color='red', linestyle='--')
axs[0, 1].set_title(r'$\lim_{x \to 0} \frac{1 - \cos x}{x} = 0$')

# Lim 3: lim x→0 (1 - cos(x))/x^2 = 1/2
axs[0, 2].plot(x, (1 - np.cos(x))/x**2)
axs[0, 2].axhline(0.5, color='red', linestyle='--')
axs[0, 2].set_title(r'$\lim_{x \to 0} \frac{1 - \cos x}{x^2} = \frac{1}{2}$')

# Lim 4: lim x→0 (tan(x)/x) = 1
axs[1, 0].plot(x, np.tan(x)/x)
axs[1, 0].axhline(1, color='red', linestyle='--')
axs[1, 0].set_title(r'$\lim_{x \to 0} \frac{\tan x}{x} = 1$')

# Lim 5: lim x→0 (e^x - 1)/x = 1
axs[1, 1].plot(x, (np.exp(x)-1)/x)
axs[1, 1].axhline(1, color='red', linestyle='--')
axs[1, 1].set_title(r'$\lim_{x \to 0} \frac{e^x - 1}{x} = 1$')

# Lim 6: lim x→0 ln(1 + x)/x = 1
axs[1, 2].plot(x, np.log(1 + x)/x)
axs[1, 2].axhline(1, color='red', linestyle='--')
axs[1, 2].set_title(r'$\lim_{x \to 0} \frac{\ln(1+x)}{x} = 1$')

# Lim 7: lim x→∞ (1 + 1/x)^x = e
axs[2, 0].plot(x_inf, (1 + 1/x_inf)**x_inf)
axs[2, 0].axhline(np.e, color='red', linestyle='--')
axs[2, 0].set_title(r'$\lim_{x \to \infty} \left(1 + \frac{1}{x}\right)^x = e$')

# Lim 8: lim x→0 (1 + x)^{1/x} = e
axs[2, 1].plot(x_pos, (1 + x_pos)**(1/x_pos))
axs[2, 1].axhline(np.e, color='red', linestyle='--')
axs[2, 1].set_title(r'$\lim_{x \to 0} \left(1 + x\right)^{\frac{1}{x}} = e$')

# Lim 9: vacía por ahora
axs[2, 2].axis('off')

# Ajuste
for ax in axs.flat:
    ax.grid(True)
    ax.set_xlim(left=min(ax.get_xlim()[0], -0.1), right=max(ax.get_xlim()[1], 0.1))

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Verificación gráfica de límites con funciones por partes", fontsize=16)

# a) f(x) con a = 3
x1_left = np.linspace(0.1, 3, 400)
x1_right = np.linspace(3.01, 4, 400)
f1_left = np.cbrt(1 - x1_left**2)
f1_right = 2 / (3 - x1_right)

axs[0].plot(x1_left, f1_left, label=r'$f(x)=\sqrt[3]{1 - x^2}$', color='blue')
axs[0].plot(x1_right, f1_right, label=r'$f(x)=\frac{2}{3 - x}$', color='green')
axs[0].plot(3, 10, 'ro', label=r'$f(3)=10$')  # punto definido en x=3
axs[0].axvline(3, color='gray', linestyle='--', alpha=0.5)
axs[0].set_title(r'a) $f(x)$ con $a=3$')
axs[0].legend()
axs[0].grid(True)

# b) f(x) con a = 0
x2_left = np.linspace(-2, -0.001, 500)
x2_right = np.linspace(0.001, 2, 500)
f2_left = (np.exp(x2_left) - x2_left - 1) / x2_left
f2_right = np.sin(x2_right**2) / x2_right

axs[1].plot(x2_left, f2_left, label=r'$f(x)=\frac{e^x - x - 1}{x}$', color='purple')
axs[1].plot(x2_right, f2_right, label=r'$f(x)=\frac{\sin(x^2)}{x}$', color='orange')
axs[1].axvline(0, color='gray', linestyle='--', alpha=0.5)
axs[1].set_title(r'b) $f(x)$ con $a=0$')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Configuración general para usar simbólicamente si es necesario
x, y, z = sp.symbols('x y z')

# Lista de funciones y puntos
ejercicios = {
    "a": (lambda x: 3, -1, r"$f(x) = 3$"),
    "b": (lambda x: x**2 - 7*x + 12, 0, r"$g(x) = x^2 - 7x + 12$"),
    "c": (lambda x: np.sqrt(x - 3), -3, r"$f(x) = \sqrt{x - 3}$"),
    "d": (lambda y: (y**2 - 1)/(5*y - 5), 1, r"$h(y) = \frac{y^2 - 1}{5y - 5}$"),
    "e": (lambda x: 3 / (2*(x - 1)**2), 0.5, r"$f(x) = \frac{3}{2(x - 1)^2}$"),
    "f": (lambda z: (np.sqrt(z + 5) - 2)/(z - 7), 7, r"$g(z) = \frac{\sqrt{z+5} - 2}{z - 7}$"),
    "g": (lambda x: (6*x + 3)/(x - 2), 2, r"$h(x) = \frac{6x + 3}{x - 2}$"),
    "h": (lambda x: (x + 2)/(x**2 - 4), [-2, 2], r"$g(x) = \frac{x + 2}{x^2 - 4}$")
}

# Crear subplots
fig, axs = plt.subplots(4, 2, figsize=(15, 18))
fig.suptitle("Ejercicio 11 y 12: Gráficos y continuidad", fontsize=18)

# Graficar cada función
for idx, (key, (func, a_val, label)) in enumerate(ejercicios.items()):
    row, col = divmod(idx, 2)
    ax = axs[row, col]

    # Definir dominio evitando singularidades
    if isinstance(a_val, list):
        a_points = a_val
    else:
        a_points = [a_val]

    x_vals = np.linspace(-10, 10, 1000)
    for a in a_points:
        x_vals = x_vals[np.abs(x_vals - a) > 1e-3]  # evitar divisiones por 0

    # Evaluar función con try-except por seguridad
    try:
        y_vals = func(x_vals)
        ax.plot(x_vals, y_vals, label=label)
        for a in a_points:
            ax.axvline(a, color='red', linestyle='--', alpha=0.5)
    except Exception as e:
        ax.text(0.5, 0.5, f"Error en función {key}", fontsize=12)

    ax.set_title(f"{key}) {label} en a={a_val}")
    ax.legend()
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp


x = sp.Symbol('x')


funciones = {
    'a': 3/(x+2),
    'b': (x**2 + 3)/(x**2 - 9),
    'c': (x**3 + 1)/(x**2 - 1),
    'd': (x - 2)/(x**2 + 2*x - 3)
}


fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

for i, (letra, f_expr) in enumerate(funciones.items()):
    # Convertir a función numérica
    f_lambdified = sp.lambdify(x, f_expr, modules=["numpy"])

    # Detectar discontinuidades
    discontinuidades = sp.singularities(f_expr, x)
    asintotas = [float(p) for p in discontinuidades if p.is_real]

    # Generar dominio evitando las asíntotas
    x_vals = np.linspace(-10, 10, 1000)
    for a in asintotas:
        x_vals = x_vals[np.abs(x_vals - a) > 0.01]  # evitar cercanías a las asíntotas

    y_vals = f_lambdified(x_vals)

    # Graficar
    axs[i].plot(x_vals, y_vals, label=f'{letra}) {str(f_expr)}')

    for a in asintotas:
        axs[i].axvline(x=a, color='red', linestyle='--', label=f'Asíntota x={a}')

    axs[i].set_ylim(-20, 20)
    axs[i].grid(True)
    axs[i].set_title(f'{letra}) f(x)')
    axs[i].legend()

plt.tight_layout()
plt.show()
```
